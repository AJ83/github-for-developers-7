## <strong>Introducción</strong>
</div>

El documento que estás leyendo no es una página web estática, sino un entorno interactivo denominado <strong>notebook de Colab</strong>, que permite escribir y ejecutar código.

Por ejemplo, esta es una <strong>celda de código</strong> con una secuencia de comandos Python corta que calcula un valor, lo almacena en una variable y devuelve el resultado:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

## Ciencia de datos
</div>

Con Colab, puedes aprovechar por completo las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código que se incluye a continuación usa <strong>NumPy</strong> para generar algunos datos aleatorios y <strong>matplotlib</strong> para visualizarlos. Para editar el código, haz clic en la celda y comienza a editar.

Puedes importar datos propios a notebooks de Colab desde tu cuenta de Google Drive &#40;incluso desde hojas de cálculos&#41;, GitHub y muchas otras fuentes. Para obtener más información acerca de la importación de datos y cómo puede usarse Colab para fines relacionados con la ciencia de datos, consulta los vínculos de <a href="#working-with-data">Cómo trabajar con datos</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# ==========================================================
# INVESTIGACIÓN DE OPERACIONES
# PROGRAMACIÓN LINEAL - MÉTODO GRÁFICO
# CASO: ASEGURADORA
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("="*60)
print("VERIFICACIÓN DEL MODELO DE PROGRAMACIÓN LINEAL")
print("="*60)

# ==========================================================
# DATOS DEL MODELO
# ==========================================================

print("\nRESTRICCIONES:")

print("R1: 20X1 + 10X2 <= 480")
print("R2: X2 >= 10")
print("R3: X2 <= 25")
print("R4: X1 >= 0")
print("R5: X2 >= 0")

print("\nFUNCIÓN OBJETIVO:")
print("Max Z = X1 + X2")

# ==========================================================
# PASO 1 - PUNTOS DE CORTE
# ==========================================================

print("\n" + "="*60)
print("PASO 1 - PUNTOS DE CORTE")
print("="*60)

# Corte con eje X1
x1_corte = 480 / 20
p1 = (x1_corte,0)

# Corte con eje X2
x2_corte = 480 / 10
p2 = (0,x2_corte)

print(f"P1 = {p1}")
print(f"P2 = {p2}")

# ==========================================================
# PASO 2 - VÉRTICES
# ==========================================================

print("\n" + "="*60)
print("PASO 2 - VÉRTICES")
print("="*60)

V1 = (0,10)
V2 = (0,25)

# Intersección R1 con X2=25
x1_v3 = (480 - 10*25)/20
V3 = (x1_v3,25)

# Intersección R1 con X2=10
x1_v4 = (480 - 10*10)/20
V4 = (x1_v4,10)

vertices = [V1,V2,V3,V4]

for i,v in enumerate(vertices,start=1):
    print(f"V{i} = {v}")

# ==========================================================
# PASO 3 - EVALUACIÓN DE LA FUNCIÓN OBJETIVO
# ==========================================================

print("\n" + "="*60)
print("PASO 3 - EVALUACIÓN DE Z")
print("="*60)

resultados = []

for i,v in enumerate(vertices,start=1):

    x1 = v[0]
    x2 = v[1]

    z = x1 + x2

    resultados.append(
        [f"V{i}",x1,x2,z]
    )

tabla = pd.DataFrame(
    resultados,
    columns=["Vértice","X1","X2","Z"]
)

print(tabla)

# ==========================================================
# PASO 4 - SOLUCIÓN ÓPTIMA
# ==========================================================

indice = tabla["Z"].idxmax()

optimo = tabla.loc[indice]

print("\n" + "="*60)
print("SOLUCIÓN ÓPTIMA")
print("="*60)

print(f"X1* = {optimo['X1']}")
print(f"X2* = {optimo['X2']}")
print(f"Zmáx = {optimo['Z']}")

# ==========================================================
# PASO 5 - VERIFICACIÓN DE RESTRICCIONES
# ==========================================================

print("\n" + "="*60)
print("VERIFICACIÓN DEL VÉRTICE ÓPTIMO")
print("="*60)

x1 = optimo["X1"]
x2 = optimo["X2"]

print("R1:",20*x1 + 10*x2,"<= 480")
print("R2:",x2,">= 10")
print("R3:",x2,"<= 25")
print("R4:",x1,">= 0")
print("R5:",x2,">= 0")

# ==========================================================
# PASO 6 - GRÁFICA
# ==========================================================

x = np.linspace(0,25,400)

y = (480 - 20*x)/10

plt.figure(figsize=(10,7))

# Recta principal
plt.plot(
    x,
    y,
    label='20X1 + 10X2 = 480'
)

# Restricciones horizontales
plt.axhline(
    y=10,
    linestyle='--',
    label='X2 = 10'
)

plt.axhline(
    y=25,
    linestyle='--',
    label='X2 = 25'
)

# Región factible
region_x = [0,0,11.5,19]
region_y = [10,25,25,10]

plt.fill(
    region_x,
    region_y,
    alpha=0.3,
    label='Región Factible'
)

# Vértices
for nombre,v in zip(
    ["V1","V2","V3","V4"],
    vertices
):

    plt.scatter(v[0],v[1],s=80)

    plt.text(
        v[0]+0.2,
        v[1]+0.5,
        nombre
    )

plt.xlabel("X1")
plt.ylabel("X2")

plt.title(
    "Método Gráfico - Caso Aseguradora"
)

plt.xlim(0,25)
plt.ylim(0,50)

plt.grid(True)

plt.legend()

plt.show()
```
